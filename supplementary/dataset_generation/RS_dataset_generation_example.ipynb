```python
from RS_dataset_generation import generate_and_calculate_signed_distance
from RS_dataset_generation import plot_example_polygon
from RS_dataset_generation import plot_sdf_surface
from RS_dataset_generation import save_dataset
import os
```

```python
# ploting geometry exmaples
filename = '../src/figures/example_polygon.png'
plot_example_polygon(filename=filename)
```

```python
# ploting SDF surface
text_formula = r"$L_{orgz}$"

# quadrangle
filename = '../src/figures/sdf_surface_quadrangle.png'
X, Y, Z, rounded_quadrangles, desired_angle_offset_sum = generate_and_calculate_signed_distance(num_vertices=4, noise_scale_z=0.02)
title = f'Quadrangle, {text_formula}: {desired_angle_offset_sum:.2f}, noise_scale_z: 0.02'
plot_sdf_surface(X, Y, Z, rounded_quadrangles, title=title, filename=filename)

# triangle
filename = '../src/figures/sdf_surface_triangle.png'
X, Y, Z, rounded_quadrangles, desired_angle_offset_sum = generate_and_calculate_signed_distance(num_vertices=3, noise_scale_z=0.02)
title = f'Triangle, {text_formula}: {desired_angle_offset_sum:.2f}, noise_scale_z: 0.02'
plot_sdf_surface(X, Y, Z, rounded_quadrangles, title=title, filename=filename)
```

```python
# dataset generation
root_dir = '../shape_datasets/RS_loss_datasets'
sub_dirs = ['triangle', 'quadrangle']

for sub_dir in sub_dirs:
    if not os.path.exists(f'{root_dir}/{sub_dir}'):
        os.makedirs(f'{root_dir}/{sub_dir}')

    num_vertices = 3 if sub_dir == 'triangle' else 4
    save_dataset(f'{root_dir}/{sub_dir}', num_samples=50000, num_vertices=num_vertices, points_per_side=64, noise_scale_z=0.02)
```
