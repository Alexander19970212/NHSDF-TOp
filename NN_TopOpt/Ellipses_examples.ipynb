```python
import sys
import os
import json

sys.path.append(os.path.abspath('..'))
sys.path.append(os.path.abspath('../models'))
sys.path.append(os.path.abspath('../dataset_generation'))

from NN_TopOpt import FeatureMappingDecSDF, FeatureMappingTopOpt, SIMP_basic
from TopOpt import TopOptimizer2D
import yaml

methods = {
    "fm_nn_heaviside": FeatureMappingTopOpt,
}
```

```python
import torch 
torch.cuda.empty_cache()

experiment_dir = "../configs/NN_top_optimization/Gaussians_Heaviside"

config_name = "MBB_beam_half" #
# config_name = "Cant_beam"
# config_name = "Cant_beam_distrL"
# config_name = "Brecket"

# config_name = "Square_beam"
# config_name = "Square_cantilever_beam"

config_path = f"{experiment_dir}/{config_name}.yaml"

with open(config_path, 'r') as file:
    config = yaml.safe_load(file)

# print(config)

params = config['params']
print(params["compliance_w"])
params['refactoring_markers'] = list(range(38, 80, 5))
# params['treshold'] = 0.5
print(params['refactoring_markers'])
TopOptimizer = TopOptimizer2D(methods, params)
TopOptimizer.optimize(plot_interval=20)
TopOptimizer.print_metrics()
```

```python
# plot geometry
geometry_features = TopOptimizer.method.gaussian_core.get_geometry()

gray = False

root_dir = "../src/model_comparison/Ellipses"

os.makedirs(root_dir, exist_ok=True)

if gray:
    filename = f"{root_dir}/{config_name}_Ellipses_gray.png"
    TopOptimizer.plot_final_result(geometry_features, filename=filename, plot_von_mises=not gray)
else:
    filename = f"{root_dir}/{config_name}.png"
    TopOptimizer.plot_final_result(geometry_features, filename=filename, plot_von_mises=True)

vf, compliance, max_stress = TopOptimizer.print_metrics()
metrics = {
    "vf": vf,
    "compliance": compliance,
    "max_stress": max_stress
}
filename_metrics = f"{root_dir}/{config_name}.json"
with open(filename_metrics, "w") as f:
    json.dump(metrics, f)
```
